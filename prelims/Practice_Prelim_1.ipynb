### 1) (24 points)
a) What was the average settled water turbidity at the Ithaca WTP averaged over all of 2018?

b) What fraction of the influent turbidity was removed by sedimentation at the Ithaca WTP averaged over all of 2018?

c) What was the average membrane filtered water turbidity at the Ithaca WTP averaged over all of 2018?

d) What fraction of the influent turbidity was removed by the membrane filters at the Ithaca WTP averaged over all of 2018?

e) What was the average turbidity of the waste from the membrane filters?

f) What was the average turbidity of the waste from the sedimentation tanks?


```
datafile = 'https://github.com/monroews/CEE4590/raw/master/DC/Ithaca_WTP_data_2018.xls'
df = pd.read_excel(datafile,sheet_name=0)

# use the next line of code to see the very long list of column headers. Comment it out when not needed!
# print(df.columns)

# get the turbidity data
Ave_raw_turbidity = np.average(df['Raw (NTU)'])  * u.NTU
Ave_settled_turbidity = np.average(df['Settled Water (NTU)'])  * u.NTU
Ave_filtered_turbidity = np.average(df['Clearwell Effluent (NTU)'])  * u.NTU
print('a) The average settled water turbidity was',Ave_settled_turbidity)

# calculate the fractions removed by each process
sed_fraction = (Ave_raw_turbidity - Ave_settled_turbidity)/Ave_raw_turbidity
fi_fraction = (Ave_settled_turbidity - Ave_filtered_turbidity)/Ave_raw_turbidity

print('b) The fraction of particles removed by the sedimentation tanks was',sed_fraction)
print('c) The fraction of particles removed by the membrane filters was',fi_fraction)

# calculate average turbidity of membrane backwash water
# first find the total mass of turbiidty in the wash water

Q_raw = ((np.sum(df['Raw Water (MGD)'])  * 10**6 * u.gal)/u.year).to(u.L/u.s)
Q_filtered = ((np.sum(df['Skid Total filtrate (MG)'])  * 10**6 * u.gal)/u.year).to(u.L/u.s)
Q_membrane_waste = ((np.sum(df['Skid Total waste (kgal)'])  * 10**3 * u.gal)/u.year).to(u.L/u.s)

Backwash_turbidity = (Ave_settled_turbidity - Ave_filtered_turbidity) * Q_filtered/Q_membrane_waste
print('d) The average backwash turbidity was', Backwash_turbidity)

Q_sed_waste = Q_raw - Q_filtered - Q_membrane_waste
Q_sed = Q_raw - Q_sed_waste
Sed_sludge_turbidity = (Ave_raw_turbidity- Ave_settled_turbidity) * Q_sed/Q_sed_waste
print('e) The settled sludge turbidity was', Sed_sludge_turbidity)
```
